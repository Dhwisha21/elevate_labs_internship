```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:/Users/Dhwisha/Downloads/netflix_movies_detailed_up_to_2025.csv")
df.head(5)
```

```python
# check missing values
df.isnull().sum()
```

```python
# drop duration column
cleaned_data=df.drop(columns=['duration'])
cleaned_data.head(5)

```

```python
# remove duplicate values
df = cleaned_data.drop_duplicates()
df.head(5)
```

```python
# replace missing values with unknown
columns_to_fill = ['director', 'cast', 'country', 'genres', 'description']
cleaned_data[columns_to_fill] = cleaned_data[columns_to_fill].fillna('Unknown')
cleaned_data.isnull().sum()
```

```python
# standardize the data
cleaned_data.columns = cleaned_data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# convert to datetime format
cleaned_data['date_added'] = pd.to_datetime(cleaned_data['date_added'], errors='coerce')
cleaned_data.dtypes
```

```python
summary = {
    "missing_values": cleaned_data.isnull().sum(),
    "duplicates_removed": df.duplicated().sum(),
    "final_shape": cleaned_data.shape,
    "check data types":cleaned_data.dtypes,
    
}

# Save cleaned dataset for GitHub structure
cleaned_path = "/Users/Dhwisha/Downloads/netflix_cleaned.csv"
cleaned_data.to_csv(cleaned_path, index=False)

summary

```
